### MAXIMUM LIKELIHOOD ESTIMATION

- We know the functional form of the distribution but the parameters of the function are uknown.
    - Eg. The data is known to come from a normal distribution but $\mu$ or $\Sigma$ or both are unknown.
    - A set of observations from a coin toss experiment are known to come from a distribution $\{ p , 1-p \}$ for H and T respectively but $p$ is unknown.
- These parameters are to estimated from the data.


#### Data Assumptions
- Denote the observations by $X = \{x_1,x_2, \cdots x_n \} $
- The individual observations are assumed to be **statistically independent of each other**. This is an assumption that is valid in most of the cases. For instance, the outcome of a coin toss has no say in the outcomes of the tosses before or after that.(One who tends to be extremely rigorous might argue that each toss causes some wear and tear in the coin and hence influences the outcomes of the subsequent tosses. This is ideally the case, but results in a very complicated picture almost impossible to be modelled)
- The second assumption is that each of the observations come from the same distribution. That is, the statistical distribution model that we have assumed for the data is valid for all the observations. In statistics parlance we say that the observations are **identically distributed**.
- In short we usually say the data is ***iid*** (independent and identically distributed).

- Let us denote the parameters of the model by $\theta$
- To reflect the fact that the probability of the observations depends on the parameter $\theta$ we write it as
$p(X;\theta)$.(It is called the likelihood of the data given the parameters $\theta$).
- ***independence assumption*** enables simplification of $p(X;\theta)$ as:
- $p(X;\theta) = p(x_1;\theta)p(x_2;\theta) \cdots p(x_n;\theta) = \Pi_{i=1}^{n} p(x_i;\theta)$
- ***identical distribution*** assumption implies that $p(x_i;\theta) \; \forall i = 1 \cdots n $ have the same functional form and parameters $\theta$


- For instance, for the coin toss experiment each observation $x_i \in \{H,T \}$ with unknown probability $p$ and $1-p$ respectively. This along with the iid assumption lead to the following relation for $p(X;\theta)$<br\> $p(X;\theta) = p^{n_H}(1-p)^{n-n_H}$.
- $\theta = p $ in this case since there is only one parameter for the distribution.

- $p(X;\theta) $ is called the likelihood function ( likelihood of data observtions given the parameters)
- Maximum Likelihood Estimation works by finding a parameter $\theta$ that maximizes the likelihood. That is,
- $\hat{\theta} =  \arg \max_{\theta} p(X;\theta) $.
- Since most of the distributions are exponential in nature generally we work with log likelihood that is $\log(p(x;\theta))$. <br\> Therefore $\hat{\theta} = \arg \max_{\theta} \log p(X;\theta) $.<br\> Log likelihood is a very important term in statistics and machine learning and we are going to come across this at many places.

- Coming back to the coin tossing experiment the log likelihood is $n_H \log p + (n-n_H) log(1-p) $
- ML estimate of $p$ is obtained by optimizing the above function wrt to p. This is easily solvable in this case. You get the solution by equating the derivative of log likelihood wrt $p$ to $0$. This yields $p = \frac{n_H}{n}$

##### Estimating the mean of a gaussian distribution.
- Estimation of the mean $\mu$ when covariance matrix $\Sigma$ is known. That is the unknown parameter $\theta$ in this case is $\mu$.
- Assume the data is generated according to normal distribution with mean $\mu$ and covariance matrix $\Sigma$
- Let $ X = \{x_1,x_2, \cdots x_n \} $ be data observations from the distribution. The likelihood function with iid assumption on the data is given by $p(X;\theta) = \prod_{i=1}^{n}  \frac{1}{(2\pi)^{n/2} |\Sigma|^n } exp(-(\frac{1}{2}(x_i-\mu)\Sigma^{-1})(x_i-\mu))$. Taking the log we get the log likelihood given by <br\> -$\frac{1}{2}\sum_{i=1}^{n}(x_i-\mu)\Sigma^{-1})(x_i-\mu) + const $. 
- Differentiating this expression with respect to $\mu$ we get $ \Sigma^{-1} (\frac{1}{n}\sum_{i}^{n}x_i - \mu) = 0 $. If the covariance matrix is invertible then this is possible only if $ \frac{1}{n} \sum_{i}^{n}x_i = \mu $.    

- This implies that the sample average is an estimate of the mean of the distribution.

#### Regression  and mle
- In classification we were trying to predict class label given an observation $x$. Observe that the class label is a discrete variable which takes only finite number of values.
- But some times the variable to be predicted can a continuous one like in the case of predicting the fat percentage from the body mass index. Another example is predicting the gas mileage from the driving speed.
- Observe that the predicted variables are continuous in nature unlike class labels.
- In regression we try to predict a dependent variable/variables ($y$) given the independent variables ($x$).
- We assume an analytical model that relates $y$ to $x$, i.e., we assume that $y=f(x;w)$ where $w$ represents the parameters of the model.

- Given iid observations we have to estimate the parameters $w$ of the model. Note that the model is deterministic as long as the all the observations fit the model exactly. But then this cannot be case as all observations are corrupted to some extent by noise. That is, what is actually observed are actually corrupted instances of $y$. In other words we say that the observations satisfy $y = f(x;w) + \eta $ where $\eta$ represents the noise. 
- Noise spoils the party by bringing in randomness into the model as there is no way in which a noise can be modelled deterministically. Now the deterministic picture changes. We need to accomodate the randomness of the noise also. 

- For this we assume that the noise follows certain probability distribution. For instance, we may assume that $\eta \approx N(0,\sigma) $ that is it comes from a normal distribution with mean 0 and variance $\sigma^2$.
- Let $X = \{x_1, x_2 , \cdots x_n \}$ be the observations of independent variable along with their $Y$ values given by $Y = \{y_1, y_2 , \cdots y_n \}$. The model assumptions lets us to the following relation between $x_i $ and $y_i$, <br\> $ y_i = f(x_i,w) + \eta_i $ where $\eta_i$ is an instance of $N(0,\sigma^2)$

- Observe that given $x$, $y$ is a normal random variable with mean $f(x;w)$ and variance $\sigma^2$. That is,$p(y|x) = \frac{1}{2\pi \sigma^2}exp(-\frac{(y-f(x;w))^2}{2\sigma^2})$
- Therefore the the likelihood of the observations $p(Y|X;w) = \prod_{i=1}^{n} p(y_i|x_i;w)$. (Here we make use of the assumption that given $y_i$s are independent of each other given their corresponding $x_i$s). With these considerations the log likelihood gets reduced to the following expression :<br\> $\log p(Y|X;w) = \frac{\sum_{i=1}{n}(y_i - f(x_i;w))^2}{2\sigma^2} - \log(2\pi \sigma^2)$. 
- The maximum likelihood estimate of $w$ is defined as : <br\> $\hat{w} = \arg \max_{w}\frac{\sum_{i=1}^{n}(y_i - f(x_i;w))^2}{2\sigma^2} - \log(2\pi \sigma^2) $

##### linear regression
- Models the dependent variable $y$ as a linear function of $x$, i.e., $f(x;w) = w^Tx$
- plugging this into the relation for ML estimate of w we get :<br\>
<br\> $\hat{w} = \arg \max_{w}\frac{\sum_{i=1}^{n}(y_i - w^Tx_i)^2}{2\sigma^2} - \log(2\pi \sigma^2) $



Taking the derivative wrt $w$ and equating to 0 ultimately leads to $w = (XX^T)^{-1}XY$ where $X$ is the $d \times n$ matrix of independent variables and $Y$ is the vector of dependent variables. $(XX^T)^{-1}X$  is called the **pseudo-inverse** of $X$.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D



xrange = np.linspace(-10,10,100)
yrange = np.linspace(-10,10,100)
x,y = np.meshgrid(xrange, yrange)

w = np.array([1,1])
z = w[0]*x + w[1]*y
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(x,y,z,alpha = 0.2, color = 'b')

x_rand = np.random.uniform(-10,10,100)
y_rand = np.random.uniform(-10,10,100)
z_noisy = w[0]*x_rand + w[1]*y_rand + 0.5*np.random.randn(100)
color = ['r' if zi < 0 else 'c' for zi in z_noisy.ravel()]
plt3d.scatter(x_rand,y_rand,z_noisy, marker='o', color = color)

plt3d.plot(2*xrange,np.zeros(100),np.zeros(100), color = 'k')
plt3d.plot(np.zeros(100),yrange,np.zeros(100), color = 'k')
plt3d.plot(np.zeros(100),np.zeros(100),2*xrange, color = 'k')

plt.show()
    
```

## Maximum Aposteriori Estimation
- MLE  treats estimation as a function optimization over $\theta$. The problem with this approach is that it relies entirely on the data alone in the estimation process. But this can lead to improper estimates as illustrated in the following instance. Suppose you have a coin that is known to be meticulously manufactured so that it is known to be almost fair. That is, $p \approx 0.5 $. In an attempt to estimate the coin you toss it N times and observe that there are nearly 60% heads. MLE will blindly conclude that $p \approx 0.6$. What is missing in this is that your level of confidence in the fairness of the coin gets no consideration in MLE. Our next attempt is therefore to try and incorporate our prior knowledge about the system. In other words we try to include our belief about the parameter p.

- Our prior knowledge about the system is incorporated in the form of a distribution over $\theta$. The major conceptual shift in this case is that $\theta$ is not a simple parameter but a random variable with its own distribution denoted by $p(\theta)$. For instance we can incorporate our prior belief about the fairness of the coin by defining $p(\theta)$ to peak at around 0.5. Here the $prob(H) = \theta$ is a random variable.

- Now we have two considerations 
     - likelihood of X given $\theta$ expressed as $p(X|\theta)$
     - our prior knowledge about $\theta$ expressed through $p(\theta)$
- Therefore the objective to be maximized turns out to be $p(X|\theta)p(\theta)$
- This approach is called **Maximum Aposteriori (MAP) Estimation **

The MAP estimate can be written as : 
    \begin{equation} \theta_{map} = \arg \max p(X|\theta)p(\theta) \end{equation}
But $ p(X|\theta)p(\theta) \propto p(\theta | X) $. Therefore MAP estimate can be interpreted as maximizing the posterior distribution of the parameter given the set of observations $X$ and hence the name Max. Aposteriori Estimation.
- What we are trying to do is to see how much the data $X$ reinforces our belief about the system.
- For practical purposes we work with the log of $p(\theta | X)$ or $log(p(X|\theta)) + \log p(\theta)$

- As an eample let us find the MAP estimate of the mean of a normal distribution assuming a prior $p(\mu) \sim N(\mu_0,\sigma_\mu)$.
- It is clear that $log (p(X | \theta) = -\frac{\sum_i(x_i - \mu)^2}{2 \sigma^2} +  \frac{(\mu - \mu_0)^2}{2 \sigma_\mu^2} + $ constant
- Differentiating and equating to zero will eventually lead to the solution given by <br\>
  \begin{equation} \mu_{map} = \frac{ \frac{\sigma_\mu^2}{\sigma^2} \sum_i x_i + \mu_0 }{n \frac{\sigma_\mu^2}{\sigma^2} +1} \end{equation}
